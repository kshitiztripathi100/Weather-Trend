## Weather Trends Analysis Global vs NewYork

```python
#importing the important python library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sb
from matplotlib import style
sb.set()
```

```python
#importing the csv file of data
df=pd.read_csv('results.csv')
df.head(2)
```

```python
# getting information about datasheet
df.info()
```

```python
#getting the mean of city_avg_temp
df['city_avg_temp'].mean()
```

```python
#replacing the null value in city_avg_temp with mean value of city_avg_temp
df['city_avg_temp'].fillna(df['city_avg_temp'].mean(),inplace=True)
```

```python
df.head(15)
```

```python
# ploting the normally distributed temperature with overall years.
world_temp_mean = df.world_avg_temp
year = df.year
plt.plot(year,world_temp_mean)
plt.title('temp vs year')
plt.xlabel('year')
plt.ylabel('world_avg_temp')
plt.show()
```

```python
#Moving average for global & city data
world_MV_avg = df.world_avg_temp.rolling(8).mean()
city_MV_avg = df.city_avg_temp.rolling(8).mean()
```

```python
#ploting the world moving average temperature
plt.plot(df['year'],world_MV_avg,label='World')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature °C") 
plt.title("World Moving Average Temperature")
plt.show()
```

```python
#ploting the moving average for the world vs city
plt.plot(df['year'],world_MV_avg,label='World')
plt.plot(df['year'],city_MV_avg,label='City')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature °C") 
plt.title("New York City Moving Average Temperature VS Rest of World")
plt.show()
```

```python
from scipy.stats import pearsonr
```

```python

# prepare data
data1 = df['city_avg_temp']
data2 = df['world_avg_temp']
# calculate Pearson's correlation
corr, _ = pearsonr(data1, data2)
print('Pearsons correlation: %.3f' % corr)
```
